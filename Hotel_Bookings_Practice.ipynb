```python
hotel['company'].value_counts()
```

```python
hotel = hotel[~hotel['country'].isnull()]
hotel = hotel.drop(columns=['company'], errors='ignore')
```

```python
hotel['agent'] = hotel['agent'].fillna(hotel['agent'].median())
```

```python
hotel = hotel[~hotel['children'].isnull()]
```

```python
hotel.isnull().sum()
```

```python
hotel.duplicated().sum()
```

```python
hotel = hotel.drop_duplicates()
```

```python
hotel['arrival_date_month'].value_counts()
```

```python
hotel['arrival_date_month'] = hotel['arrival_date_month'].replace({'August':8,'July':7,'May':5,'April':4,'June':6,'March':3,'October':10,'September':9,'February':2,'December':12,'November':11,'January':1})
```

```python
hotel['arrival_date_month']
```

```python
hotel['arrival_date_month'] = hotel['arrival_date_month'].astype('str')
hotel['arrival_date_day_of_month'] = hotel['arrival_date_day_of_month'].astype('str')
hotel['arrival_date_year'] = hotel['arrival_date_year'].astype('str')
```

```python
hotel['arrival_date'] = hotel['arrival_date_month'].astype(str)+'/'+hotel['arrival_date_day_of_month'].astype(str)+'/'+hotel['arrival_date_year'].astype(str)
```

```python
hotel['arrival_date_month'] = hotel['arrival_date_month'].astype('int')
hotel['arrival_date_day_of_month'] = hotel['arrival_date_day_of_month'].astype('int')
hotel['arrival_date_year'] = hotel['arrival_date_year'].astype('int')
```

```python
hotel['arrival_date'] = pd.to_datetime(hotel['arrival_date'])
```

```python
hotel.info()
```

```python
hotel = hotel.drop(columns=['arrival_date_year','arrival_date_month','arrival_date_day_of_month'])
```

```python
columns = []
for col in hotel:
    if pd.api.types.is_numeric_dtype(hotel[col]):
        columns.append(col)
correlation_matrix = hotel[columns].corr()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title('Correlation Matrix of hotel data')
plt.show()
```
